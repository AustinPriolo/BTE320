```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
    res = num1 + num2
elif op == "-":
    res = num1 - num2
elif op == "*":
    res = num1 * num2
elif op == "/":
    res = num1 / num2
elif op == "//":
    res = num1 // num2
elif op == "**":
    res = num1 ** num2
elif op == "%":
    res = num1 % num2
else:
    print("Invalid operation")
    res = None

print(f"Result: {res}")
```

```python
import random

play = input("Do you want to play? (y/n): ").lower()

while play == "y":
    p_move = input("Choose between rock, paper, scissors: ").lower()
    c_move = random.choice(['rock', 'paper', 'scissors'])

    print(f"Player move: {p_move}")
    print(f"Computer move: {c_move}")

    if p_move == c_move:
        print("Draw")
    elif ((p_move == "rock" and c_move == "scissors") or
          (p_move == "scissors" and c_move == "paper") or
          (p_move == "paper" and c_move == "rock")):
        print("Player wins!")
    else:
        print("Computer wins!")

    play = input("Do you want to play again? (y/n): ").lower()
```

```python
def reverse(s):
    if len(s) < 2:
        return s
    else:
        #return s[-1] + reverse(s[:-1]) #take the last char and put it first
        #return reverse(s[1:]) + s[0] #take the first char and put it last
        return s[-1] + reverse(s[1:-1]) + s[0]

reverse('12345')
```

```python
def roi(revenue, cost):
    assert cost != 0
    assert isinstance(revenue, (int, float))
    assert isinstance(cost, (int, float))
    return (revenue - cost) / cost
```

```python
def loan_payment(principal, annual_rate, months):
    assert isinstance(principal, (int, float))
    assert isinstance(annual_rate, (int, float))
    assert isinstance(months, int)

    if principal <= 0 or annual_rate <= 0 or months <= 0:
        raise ValueError("Invalid input values")

    try:
        r = annual_rate / 12 / 100
        if r == 0:
            pmt = principal / months
        else:
            pmt = r * principal / (1 - (1 + r) ** (-months))
    except Exception as e:
        raise Exception(f"Error: {e}")

    return pmt
```
